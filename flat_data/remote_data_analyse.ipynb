# Summary of MCMC of Moureau et al. 2015 data set

```python
import os, re, subprocess
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

- Read the data sets and store in all_data. 
    
    - Extract the variable names and save only the ucld.mean, meanRate, and the likelihood

```python
f_names = ['Moureau_dates.log', 'Moureau_randomised10.log', 'Moureau_randomised7.log', 'Moureau_randomised3.log' ,'Moureau_randomised4.log', 'Moureau_randomised5.log', 'Moureau_randomised6.log', 'Moureau_randomised1.log', 'Moureau_randomised8.log', 'Moureau_randomised9.log', 'Moureau_randomised2.log'] 
all_data = []
titles = re.split('\t', subprocess.check_output('awk \'/state/\' Moureau_dates.log', shell = True))
for i in f_names:
    data_temp = pd.read_csv(i, sep = '\t', comment = '#', header = True, skiprows=10)
    data_temp.columns = titles
    data_temp = data_temp.ix[:, ['likelihood', 'ucld.mean', 'meanRate']]
    data_temp = data_temp.ix[3000:, ]
    all_data.append(data_temp)
```

- Make function to determine whether two hpds overlap. This is the diagnostic of the date-randomisation test

```python
def overlaps(true_vals, test_vals):
    if (max(true_vals) > max(test_vals)) and (min(true_vals) > max(test_vals)):
        return False
    elif max(true_vals) < min(test_vals) and min(true_vals) < min(test_vals):
        return False
    else:
        return True
```

- Plot the likelihood and the ucld.mean traces

```python
mean_vals = np.array([np.mean(i.ix[:, 'ucld.mean']) for i in all_data])
hpds_vals = np.array([np.percentile(i.ix[:, 'ucld.mean'], [2.5, 97.5]) for i in all_data])
hpds = np.array([np.diff(i)[0] for i in hpds_vals])
locations = [i for i in range(len(all_data))]
for d_temp in all_data:
    plt.plot(d_temp.ix[:, 'likelihood'])
plt.show()
plt.title('Trace of likelihood')

for d_temp in all_data:
    plt.plot(d_temp.ix[:, 'ucld.mean'])
plt.title('Trace of ucld.mean')
#plt.ylim((0.004, 0.012))
```

- Plot the error bars and hpd for ucld.mean estimates so far

```python
hpds_vals_list = [mean_vals - list(hpds_vals[:,0]), list(hpds_vals[:,1]) - mean_vals]
plt.errorbar(x = locations, y = mean_vals, yerr = hpds_vals_list, fmt ='o')
plt.xlim((-1.5, 12))
#plt.ylim((0.002, 0.006))
plt.ylim((0.000, 0.012))
```

- Plot histograms of ucld.mean estimates

```python
for i in range(len(all_data)):
    plt.hist(np.array(all_data[i].ix[:, 'ucld.mean']))
plt.xlim((0, 0.012))
```

- Make table with ucld.mean estimates and their uncertainty and whether they overlap with that with the correct sampling times.

```python
summary_data = pd.DataFrame(np.hstack([np.transpose(np.vstack([mean_vals, hpds])), hpds_vals]))
overlap_randomisations = pd.DataFrame([overlaps(summary_data.ix[0, [2, 3]], summary_data.ix[i, [2, 3]]) for i in range(summary_data.shape[0])])
summary_data = pd.concat([summary_data, overlap_randomisations], axis = 1, ignore_index=True)
summary_data.columns = ['mean_rate', 'hpd_width', 'lower_hpd', 'higher_hpd' , 'overlaps']
summary_data.index = f_names
summary_data
```
