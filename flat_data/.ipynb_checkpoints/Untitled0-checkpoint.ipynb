#Summary of MCMC

```python
import os, re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

## Moureau et al. 2015 data set

- Load log file:

```python
moureau_log = pd.read_csv('summ_moureau.csv')
moureau_log.head()
```

- Plot trace and histogram of likelihood removing the first 2000 steps

```python
plt.plot(np.array(moureau_log['likelihood'][2000:]))
plt.title('Likelihood trace')
```

```python
plt.hist(np.array(moureau_log['likelihood'][2000:]))
plt.title('Likelihood histogram')
```

- Plot trace and histogram for rate estimate

```python
plt.plot(np.array(moureau_log['ucld.mean'][2000:]))
plt.title('Trace of ucld mean')
```

```python
plt.hist(np.array(moureau_log['ucld.mean'][2000:]))
plt.title('Histogram of ucld mean')
```
